# Response Amplification Factor vs Frequency Ratio : 1.0

The Response Amplification Factor is also called the **Dynamic Magnification Factor.**
It is the ratio of the maximum response of the structure with the ground motion to the maximum response of the structure for some value of the damping ratio


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# import plotly.express as px
import plotly.graph_objects as go
import sympy as sp
```

```python
frequency_ratio = np.linspace(0.1, 3, 100) # range of frequency ratios; 0.1 to 3 with 100 points 
num_damping_ratios = 6 # number of damping ratios to plot
damping_ratios = np.linspace(0.01, 1.0, num_damping_ratios) # range of damping ratios; 0.01 to 1.0 with num_damping_ratios points

# loop through each damping ratio and plot the magnification factor
for damping_ratio in damping_ratios:
    mag_factor = 1 / np.sqrt((1 - frequency_ratio**2)**2 + (2 * damping_ratio * frequency_ratio)**2) # magnification factor
    plt.plot(frequency_ratio, mag_factor, label=f'Damping Ratio = {damping_ratio:.2f}') # plot magnification factor vs frequency ratio
    # plt.plot(frequency_ratio, mag_factor, label=f'Damping Ratio = {damping_ratio:.2f}', linewidth=3) # plot magnification factor vs frequency ratio with thicker line
    # plt.plot(frequency_ratio, mag_factor, label=f'Damping Ratio = {damping_ratio:.2f}', linewidth=3, linestyle='--') # plot magnification factor vs frequency ratio with thicker line and dashed line style
    # plt.plot(frequency_ratio, mag_factor, label=f'Damping Ratio = {damping_ratio:.2f}', linewidth=3, linestyle='--', marker='o') # plot magnification factor vs frequency ratio with thicker line, dashed line style, and circle markers
    # plt.plot(frequency_ratio, mag_factor, label=f'Damping Ratio = {damping_ratio:.2f}', linewidth=3, linestyle='--', marker='o', markersize=10) # plot magnification factor vs frequency ratio with thicker line, dashed line style, circle markers, and larger markers

# settings for plot by matplotlib
plt.xlabel('Frequency Ratio') # x-axis label; use r'$\omega/\omega_n$' for LaTeX formatting
plt.ylabel('Magnification Factor') # y-axis label; use r'$M$' for LaTeX formatting
plt.title(f'Magnification Factor vs Frequency Ratio for {num_damping_ratios} Damping Ratios') # title of plot; use f-string to insert num_damping_ratios into title
plt.legend() # show legend; use loc='upper right' to change location of legend; use bbox_to_anchor=(1.0, 1.0) to change location of legend; use ncol=2 to change number of columns in legend
plt.grid(True) # show grid; use linestyle='--' to change grid line style
plt.show()
```

```python
frequency_ratio = np.linspace(0.1, 3, 100)  # range of frequency ratios; 0.1 to 3 with 100 points
num_damping_ratios = 4 # number of damping ratios to plot
damping_ratios = np.linspace(0.1, 0.8, num_damping_ratios) # range of damping ratios; 0.1 to 0.8 with num_damping_ratios points


fig = go.Figure() # create a new figure; use go.FigureWidget() to create an interactive figure

# Plotting for each damping ratio
for damping_ratio in damping_ratios:
    # Magnification factor
    mag_factor = 1 / np.sqrt((1 - frequency_ratio**2)**2 + (2 * damping_ratio * frequency_ratio)**2) 
    fig.add_trace(go.Scatter(x=frequency_ratio, y=mag_factor, mode='lines', name=f'Damping Ratio = {damping_ratio:.2f}')) # plot magnification factor vs frequency ratio

# Layout settings
fig.update_layout(
    title=f'Magnification Factor vs Frequency Ratio for {num_damping_ratios} Damping Ratios',
    xaxis_title='Frequency Ratio',
    yaxis_title='Magnification Factor',
    legend=dict(title='Damping Ratio'),
    template='plotly_dark',  # Use a dark theme for a cooler appearance
)

# Show the interactive plot
fig.show()
```

```python
frequency_ratio_sym = sp.Symbol('frequency_ratio')  # Define the frequency ratio symbol
damping_ratio_sym = sp.Symbol('damping_ratio') # Define the damping ratio symbol

# --- Magnification Factor formula---
mag_factor_sym = 1 / sp.sqrt((1 - frequency_ratio_sym**2)**2 + (2 * damping_ratio_sym * frequency_ratio_sym)**2)

mag_factor_func = sp.lambdify((frequency_ratio_sym, damping_ratio_sym), mag_factor_sym, 'numpy') # Create a function for the magnification factor
# mag_factor_func = sp.lambdify((frequency_ratio_sym, damping_ratio_sym), mag_factor_sym, 'math') # Create a function for the magnification factor
# a lambda function is a small anonymous function. A lambda function can take any number of arguments, but can only have one expression.
# lambda arguments : expression 

frequency_ratio_values = np.linspace(0.1, 3, 100)  # Define the range of frequency ratios to plot
num_damping_ratios = 4 # Define the number of damping ratios to plot
damping_ratio_values = np.linspace(0.1, 0.8, num_damping_ratios) # Define the range of damping ratios to plot


fig = go.Figure() # Create a new figure; use go.FigureWidget() to create an interactive figure

# Plotting for each damping ratio
for damping_ratio in damping_ratio_values:
    mag_factor_values = mag_factor_func(frequency_ratio_values, damping_ratio)
    fig.add_trace(go.Scatter(x=frequency_ratio_values, y=mag_factor_values, mode='lines', name=f'Damping Ratio = {damping_ratio:.2f}'))

# Layout settings
fig.update_layout(
    title=f'Magnification Factor vs Frequency Ratio for {num_damping_ratios} Damping Ratios',
    xaxis_title='Frequency Ratio',
    yaxis_title='Magnification Factor ',
    legend=dict(title='Damping Ratio'),
    template='plotly_dark',  # Use a dark theme for a cooler appearance
)

# Show the interactive plot
fig.show()
```
